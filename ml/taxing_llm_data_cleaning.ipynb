```python
# trying to extract all the pdf data into text
def extract_pdf_text(file_path):
    with pdfplumber.open(pdf_path) as pdf:
        all_text=''
        for page in pdf.pages:
            page_text=page.extract_text()
            all_text+=page_text
    return all_text
pdf_path="C:/Users/mvsla/OneDrive/Desktop/LLM/taxation_data.pdf"
start_time=time.time()
try:
    text=extract_pdf_text(pdf_path)
    print('Extraction successful')
except Exception as e:
    print('Extraction failed due to:',e)
end_time=time.time()
total_time=start_time-end_time
print(f'The total time consumed for text extraction is {total_time} in seconds')
```

```python
# Function to change date format
def change_date_format(text):
    # Find all dates in the text
    dates = re.findall(r'\b\d{2}.\d{2}.\d{4}\b', text)
    mys=re.findall(r'\b\d{2}.\d{4}\b',text)         #month and year format dates
    
    # Replace each date with the new format
    for date in dates:
        new_date = datetime.strptime(date, '%d.%m.%Y').strftime('%d-%m-%Y')
        
        text = text.replace(date, new_date)
        
    return text

# Change the date format in the text data
updated_text = change_date_format(updated_text)
print(updated_text)
```
